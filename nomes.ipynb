```python
from tqdm.notebook import tqdm

male_name_set = set()
female_name_set = set()

i = 0
for row_tuple in tqdm(df_nomes.itertuples(),'loading names',df_nomes.shape[0]):
    # if i == 3:
    #     print(row_tuple.alternative_names)
    #     print(pd.isna(row_tuple.alternative_names))
    #     break
    match row_tuple.classification:
        case 'F':
            female_name_set.add(row_tuple.first_name)
            if not pd.isna(row_tuple.alternative_names):
                female_name_set.update(set(row_tuple.alternative_names.split('|')))
        case 'M':
            male_name_set.add(row_tuple.first_name)
            if not pd.isna(row_tuple.alternative_names):
                male_name_set.update(set(row_tuple.alternative_names.split('|')))
    i += 1

print(len(male_name_set))
print(len(female_name_set))
```

```python
import unicodedata

def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    return ''.join([c for c in nfkd_form if not unicodedata.combining(c)])
```

```python
def get_sex(name:str|type(pd.NA)):
    if pd.isna(name):
        return None
    first_name = name.split()[0]
    formatted_first_name = remove_accents(first_name).upper()

    is_fem = False
    is_male = False
    if formatted_first_name in female_name_set:
        is_fem = True
    if formatted_first_name in male_name_set:
        is_male = is_fem
    if first_name=='JOSE':
        print('JOSE')
        print(is_male)
        print(is_fem)
        raise ValueError('JOSE')
    
    if is_fem and is_male:
        if first_name=='JOSE':
            print('JOSE')
        mask = df_nomes['first_name']==first_name
        if first_name=='JOSE':
            print(sum(mask))
        count_hits = sum(mask)
        if count_hits > 1:
            raise ValueError(f'First name "{first_name}" was found in many rows')
        if count_hits==1:
            sex = df_nomes.loc[mask].iloc[0]['classification']
            if first_name=='JOSE':
                print(f'sex: {sex}')
            print(f'found {sex} for {first_name}')
            return df_nomes.loc[mask].iloc[0]['classification']
        if first_name=='JOSE':
            print(f'no sex found')
            raise ValueError('no sex found')
        return None
    if is_fem:
        return 'F'
    if is_male:
        return 'M'
        
    return None

def create_sex_col(author_name_col: pd.Series):
    return author_name_col.copy().apply(get_sex)
```

```python
def get_sex_ambiguous(name:str|type(pd.NA)):
    '''
    Igual a get_sex. Mas marcamos sexos decididos por desempate de first_name como M_a ou F_a
    O sufixo "_a" vem de "ambíguo"
    '''


    if pd.isna(name):
        return None
    first_name = name.split()[0]
    formatted_first_name = remove_accents(first_name).upper()

    is_fem = False
    is_male = False
    if formatted_first_name in female_name_set:
        is_fem = True
    if formatted_first_name in male_name_set:
        is_male = True
    # if formatted_first_name=='JOSE':
    #     print('JOSE')
    #     print(is_male)
    #     print(is_fem)
    #     raise ValueError('JOSE')
    
    if is_fem and is_male:
        # if formatted_first_name=='JOSE':
        #     print('JOSE')
        mask = df_nomes['first_name']==formatted_first_name
        # if formatted_first_name=='JOSE':
        #     print(sum(mask))
        count_hits = sum(mask)
        if count_hits > 1:
            raise ValueError(f'First name "{formatted_first_name}" was found in many rows')
        if count_hits==1:
            sex = df_nomes.loc[mask].iloc[0]['classification']
            # if formatted_first_name=='JOSE':
            #     print(f'sex: {sex}')
            # print(f'found {sex} for {formatted_first_name}')
            return df_nomes.loc[mask].iloc[0]['classification'] + '_a' # a for ambiguous
        if formatted_first_name=='JOSE':
            # print(f'no sex found')
            raise ValueError('no sex found')
        return None
    if is_fem:
        return 'F'
    if is_male:
        return 'M'
        
    return None

names = []
sex = []
for i in range(1,20):
    for name in df_cbot[f'autor {i}'].values:
        names.append(name)
        sex.append(get_sex_ambiguous(name))
```
